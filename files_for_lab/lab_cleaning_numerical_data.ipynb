### 6. Define a function that differentiates between continuous and discrete variables. Hint: The number of unique values might be useful. Store continuous data into a continuous_df variable and do the same for discrete_df

```python
# Checking the variability of values in the dataset.
    # High values mean more variability.
    # Lower values mean less variability.
# We will use these percentages to define whether the variable will be considered
# continuous or discrete.

customer_df.nunique()/len(customer_df)*100
```

```python
customer_df.shape[0] * 0.01 
customer_df.nunique()/len(customer_df)*100
```

```python
customer_df.shape
```

```python
# Maria and Fernando have helped me to understand how to define this function: all the credit to them. 

def cont_disc(DataFrame):
    continuous = []
    discrete = []

    threshold = customer_df.shape[0] * 0.01 

    for x in customer_df.columns:
        if customer_df[x].nunique() < threshold:
            discrete.append(x)
        else:
            continuous.append(x)
    return discrete, continuous

discrete, continuous = cont_disc(customer_df)
```

```python
discrete_df = customer_df[discrete]
discrete_df
```

```python
continuous_df = customer_df[continuous]
continuous_df
```

### 8. Create a function to plot every discrete variable. Do the same with the continuous variables (Be Careful, you may need to change the plot type to one better suited for continuous data!)

```python
def disc_plot(i):
    
    for i in discrete_df:
        sns.boxplot(x=i, y=continuous_df['total_claim_amount'], data=discrete_df)
        plt.show()
    
disc_plot(i)
```

```python
def cont_plot(i):
    
    for i in continuous_df.columns:
        sns.displot(continuous_df[i].drop['customer'], kde = True)
        plt.show()

cont_plot(i)
```
